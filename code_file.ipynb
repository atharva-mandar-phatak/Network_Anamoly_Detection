```python
from sklearn.preprocessing import OneHotEncoder
cat_features = ['protocoltype', 'service', 'flag']
encoder = OneHotEncoder(sparse_output=False)
one_hot_encoded = encoder.fit_transform(dataset[cat_features])
one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(cat_features))
dataset_encoded = pd.concat([dataset, one_hot_df], axis=1)
dataset_encoded = dataset_encoded.drop(cat_features, axis=1)
```

```python
for i in dataset_encoded.columns:
    print(i)
```

```python
dataset_encoded.columns.get_loc('service_tftp_u')
```

```python
X = dataset_encoded.drop('attack', axis= 1)
y = dataset_encoded['attack']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.21, random_state=21)
```

```python
from xgboost import XGBClassifier
# fit model no training data
xgmodel = XGBClassifier()
xgmodel.fit(X_train, y_train)
```

```python

# make predictions for test data
y_pred = xgmodel.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test,y_pred)
```
